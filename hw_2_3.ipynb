```python
author_data = data.loc[:, ["author", "name"]].groupby("author", group_keys=False).agg(book_number=('name', 'count'))
author_data
```

```python
# Якої розмірності вийшла таблиця? Відповідь: (248, 1)
```

```python
# Який автор має найбільше книг? Відповідь: Jeff Kinney
ind = author_data[author_data["book_number"] == author_data["book_number"].max()]
ind
```

```python
# Скільки книг цього автора? Відповідь: 12
```

```python
rating_data = data.loc[:, ["author", "reviews"]].groupby("author", group_keys=False).agg(author_rating=('reviews', 'mean'))
rating_data
```

```python
# У якого автора середній рейтинг мінімальний? Відповідь: Zhi Gang Sha
ind = rating_data[rating_data["author_rating"] == rating_data["author_rating"].min()]
ind
```

```python
# Який у цього автора середній рейтинг? Відповідь: 128.5
```

```python
data_1 = pd.concat([author_data, rating_data], axis=1)
data_1
```

```python
data_1 = data_1.sort_values(by=["book_number", "author_rating"], ascending=True)
data_1
```

```python
# Який автор перший у списку?
# Peter A. Lillback
```

```python
data_1 = data[data["genre"] == "Fiction"].sort_values(by=["year"], ascending=True)
data_1
```

```python
sns.histplot(data=data_1, x=data_1['price'], kde=True, bins=20)
plt.title('Розподіл цін на книги для жанру Fiction')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
data_2 = data[data["genre"] == "Non Fiction"].sort_values(by=["year"], ascending=True)
data_2
```

```python
sns.histplot(data=data, x=data_2['price'], kde=True, bins=20)
plt.title('Розподіл цін на книги для жанру Non Fiction')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=data, x=data['price'], hue='genre', kde=True, bins=20)
plt.title('Розподіл цін на книги для різних жанрів')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=data, x=data['user_rating'], y=data['reviews'], hue='genre')
plt.title('Залежність рейтингу від кількості відгуків')
plt.xlabel('Рейтинг')
plt.ylabel('Кількість відгуків')
plt.show()
```

```python
avg_price_by_year = data.groupby('year')['price'].mean()
avg_price_by_year.plot(marker='o', linestyle='-', color='b')
plt.title('Динаміка зміни цін на книги')
plt.xlabel('Рік')
plt.ylabel('Середня ціна')
plt.show()
```
