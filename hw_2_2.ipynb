```python
def fill_avg_salary(row):
    avg_salary = np.nanmean([row['Зарплата.в.месяц'], row['Изменение.зарплаты.за.12.месяцев']])
    return avg_salary 
```

```python
python_data['avg'] = python_data.apply(fill_avg_salary, axis=1)
python_data
```

```python
avg_salary_by_education = data.groupby('Образование')['salary'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_education.index, y=avg_salary_by_education.values, palette='viridis')
plt.title('Середня зарплата за рівнем освіти серед всіх розробників')
plt.xlabel('Рівень освіти')
plt.ylabel('Середня зарплата')
plt.xticks(rotation=45, ha='right')
plt.show()
```
