```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

page = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
table_name = "Коефіцієнт народжуваності у регіонах України"
data = pd.read_html(page, match=table_name, thousands=".", decimal=",")
data
```

```python
data_head = data[0].head()
data_head
```

```python
data_info = data[0].shape
print(data_info)
```

```python
data = data[0].replace(0, np.nan)
data
```

```python
print(data.dtypes)
```

```python
data_ = data.replace(np.nan, 0)
data_
```

```python
print(data.isnull().sum())
```

```python
data = data.drop(index=data.index[-1])
data
```

```python
# print(data['1990'].mean())
for i in data.columns[1:]:
    data[i].fillna(data[i].mean(), inplace=True)
data
```

```python
avg = ["avg"]
for i in data.columns[1:]:
    avg.append(data[i].mean())
print(avg)
```

```python
data.loc[len(data)] = avg
data
```

```python
mean_val = data["2014"].mean()
shape_ = data.shape
for i in range(shape_[0]):
    if data["2014"][i] > mean_val:
        print(data["регіон"][i])
print()
[print(data["регіон"][i]) for i in range(shape_[0]) if data["2014"][i] > mean_val]
```

```python
max_value = max(data["2014"])
print(max_value)
ind = data[data["2014"] == max_value].index
print(data["регіон"][ind[0]])
```

```python
# maybe more pandas version but having some issues in compare with previuos
mean_val = data["2014"].mean()

for i in data["2014"]:
    if i > mean_val:
        ind = data[data["2014"] == i].index
        print(data["регіон"][ind[0]])
```

```python
plt.figure(figsize=(12, 8))
plt.bar(data["регіон"],
       data["2014"])

plt.xticks(rotation=45, ha='right')
plt.ylabel("Number, thousands", fontsize="small", color="midnightblue")
plt.title("Birth", fontsize=15)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
x_data = [i for i in data.columns]

sns.pointplot(x=x_data[1:], y=data.iloc[0][1:], label='АР Крим')
plt.title('Зміна рівня народжуванності')
plt.xlabel('Рік')
plt.ylabel('Рівень народжуванності')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
x_data = [i for i in data.columns]

sns.lineplot(x=x_data[1:], y=data.iloc[0][1:], marker='o', label='АР Крим')
sns.lineplot(x=x_data[1:], y=data.iloc[1][1:], marker='s', label='Вінницька область')
sns.lineplot(x=x_data[1:], y=data.iloc[2][1:],  marker='^', label='Волинська область')
sns.lineplot(x=x_data[1:], y=data.iloc[4][1:],  marker='o', label='Донецька область')

plt.title('Зміна рівня народжуванності')
plt.xlabel('Рік')
plt.ylabel('Рівень народжуванності')
plt.legend()
plt.show()
```

```python
data = data.drop(index=data.index[-1])

plt.figure(figsize=(12, 8))
sns.barplot(x='регіон', y='1990', data=data)
plt.title('Рівень народжуванності в 1990 році')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуванності')
plt.xticks(rotation=45, ha='right')
plt.show()
```
